```python
from playwright.async_api import async_playwright
import asyncio

async def scrape_job_posting(job_url):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context(user_agent="Mozilla/5.0")
        page = await context.new_page()

        await page.goto(job_url)
        await asyncio.sleep(5)  # Wait for 5 seconds

        # Saving the page's HTML content for debugging purposes
        html = await page.content()
        with open("page_debug.html", "w", encoding="utf-8") as f:
            f.write(html)
        print("Saved page content to page_debug.html")

        # Taking a screenshot of the page for verification
        await page.screenshot(path="screenshot.png", full_page=True)
        print("Screenshot saved as screenshot.png")


        # Extract job title
        job_title = await page.inner_text('h1.top-card-layout__title')  # Adjust selector
        company = await page.inner_text('a.topcard__org-name-link')  # Adjust selector
        location = await page.inner_text('span.topcard__flavor--bullet')  # Adjust selector
        description = await page.inner_text('div.show-more-less-html__markup')  # Adjust selector

        print(f"Job Title: {job_title}")
        print(f"Company: {company}")
        print(f"Location: {location}")
        print(f"Description: {description}")

        await browser.close()
        return job_title, company, location, description

job_url = "https://www.linkedin.com/jobs/view/4114601385/"
job_title, company, location, description = await scrape_job_posting(job_url)
```

```python
def condense_job_description(description):
    """
    Condense the job description into a single string without unnecessary spaces or line breaks.
    """
    # Normalize whitespace by joining all lines and removing excessive spaces
    condensed_description = " ".join(description.split())
    return condensed_description
```

```python
import spacy

# Load spaCy NLP model
nlp = spacy.load("en_core_web_sm")

# Function to extract and filter skills dynamically
def extract_dynamic_skills(description):
    """
    Extract skills dynamically from a job description using heuristics and POS tagging.
    """
    # Preprocess the description
    description = " ".join(description.split())

    # Parse with spaCy
    doc = nlp(description)

    # Extract potential skills based on noun phrases
    skills = set()
    for chunk in doc.noun_chunks:
        if len(chunk.text) > 2:  # Exclude very short terms
            skills.add(chunk.text.strip().lower())

    # Filter using POS tagging
    for token in doc:
        if token.pos_ in {"NOUN", "PROPN"} and len(token.text) > 2:
            skills.add(token.text.strip().lower())

    # Deduplicate and return sorted skills
    return sorted(skills)



# Extract skills dynamically
skills = extract_dynamic_skills(description)

# Output the extracted skills
print("Extracted Skills:")
for skill in skills:
    print(f"- {skill}")
```

```python
import spacy
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

def extract_technical_phrases(job_descriptions):
    """
    Extract technical phrases from job descriptions using linguistic patterns and dependency parsing.
    """
    technical_phrases = []

    for description in job_descriptions:
        doc = nlp(description)

        # Extract noun phrases with compounds or adjectives
        for chunk in doc.noun_chunks:
            chunk_text = chunk.text.lower().strip()
            if len(chunk_text.split()) > 1 and not chunk_text.startswith(("the", "a", "an", "our")):
                technical_phrases.append(chunk_text)

        # Extract verbs and their objects for technical actions
        for token in doc:
            if token.pos_ == "VERB":
                # Find direct or prepositional objects of the verb
                objects = [child.text.lower() for child in token.children if child.dep_ in ("dobj", "pobj")]
                if objects:
                    phrase = f"{token.lemma_} {' '.join(objects)}"
                    technical_phrases.append(phrase)

        # Extract proper nouns and technical terms
        for token in doc:
            if token.pos_ in {"PROPN", "NOUN"} and not token.is_stop and len(token.text) > 1:
                technical_phrases.append(token.text.lower())

    return technical_phrases

def rank_technical_phrases(job_descriptions, max_features=50):
    """
    Rank technical phrases using TF-IDF to identify the most significant ones.
    """
    # Extract technical phrases
    technical_phrases = extract_technical_phrases(job_descriptions)

    # Use TF-IDF to rank phrases
    vectorizer = TfidfVectorizer(
        stop_words="english",
        max_features=max_features,
        ngram_range=(1, 3)
    )
    tfidf_matrix = vectorizer.fit_transform([" ".join(technical_phrases)])
    tfidf_features = vectorizer.get_feature_names_out()

    # Rank by frequency
    term_counts = Counter(technical_phrases)
    ranked_phrases = sorted(
        tfidf_features,
        key=lambda x: term_counts[x],
        reverse=True
    )

    return ranked_phrases

# Example DataFrame with job descriptions
data = {"job_descriptions": [
    "We are looking for a Data Scientist with experience in Python, SQL, and machine learning.",
    "Join our team as a Software Engineer specializing in C++, Java, and cloud computing.",
    "Camtek is hiring a Computer Vision Engineer with skills in image processing, deep learning, and algorithms.",
]}
df = pd.DataFrame(data)

# Extract and rank technical phrases
technical_skills = rank_technical_phrases(df["job_descriptions"], max_features=20)
print("Extracted Technical Skills:", technical_skills)
```

```python
#code for single description
import spacy
import re

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Function to extract skills, responsibilities, and qualifications
def extract_skills_from_string(job_description):
    # Ensure input is a string
    if not isinstance(job_description, str):
        raise ValueError("Input must be a string representing the job description.")

    # Use regex to split the job description into sections
    sections = re.split(r"\n\n+", job_description)
    skills = set()

    for section in sections:
        # Focus on sections that likely contain skills
        if any(keyword in section.lower() for keyword in ["responsibilities", "qualifications", "experience", "skills"]):
            doc = nlp(section)
            for chunk in doc.noun_chunks:
                # Include meaningful noun phrases
                skills.add(chunk.text.strip().lower())
            for token in doc:
                # Include standalone nouns and proper nouns
                if token.pos_ in {"NOUN", "PROPN"} and len(token.text) > 2:
                    skills.add(token.text.strip().lower())

    # Post-process to filter irrelevant terms
    stop_words = {"the", "and", "or", "to", "in", "of", "a", "is", "with", "on", "for", "as", "this", "must", "advantage"}
    filtered_skills = [skill for skill in skills if skill not in stop_words]

    return sorted(filtered_skills)



# Extract skills
skills = extract_skills_from_string(description)

# Output the results
print("Extracted Skills:", skills)
```
